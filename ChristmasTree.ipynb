```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
```

```python
mu, sigma = 0, 1 # mean and standard deviation
s = np.random.normal(mu, sigma, int(1e5))

def gaussian(x):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2) )
```

```python
plt.figure(dpi=300)
count, bins, ignored = plt.hist(s, 100, density=True, color='g')


x = np.logspace(-2.5, 0.9, 100, base=np.exp(1.))
x = np.concatenate([-x, x])
x = np.sort(x)

colors = pl.cm.tab10(np.linspace(0,1,len(x)))
rand = np.random.rand(len(x))
randint = np.random.randint(0, len(x), len(x))

plt.xticks([])
plt.yticks([])

for i in range(len(x)):

    plt.plot(x[i], gaussian(x[i])*rand[i], 'o', c=colors[randint[i]])

plt.show()
```
