```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge, Circle
```

```python
x = np.linspace(0., 9.5, 40)
x2 = x
```

```python
def openMouth(x):
    fig, ax = plt.subplots(figsize=(10, 3), dpi=241.635, facecolor='#404040')
    ax.set_axis_off()
    ax.set_xlim(0., 10.)
    ax.set_ylim(-0.1, 3.1)
    plt.plot([0.25, 9.75], [2.75, 2.75], '#2121DE', linewidth=2)
    plt.plot([0.25, 9.75], [3.05, 3.05], '#2121DE', linewidth=2)
    plt.plot([0.25, 9.75], [0.25, 0.25], '#2121DE', linewidth=2)   
    plt.plot([0.25, 9.75], [-0.05, -0.05], '#2121DE', linewidth=2)   
    p = Wedge((x, 1.5), 1., 45, -45, color='#FFFF00')
    ax.add_patch(p)
    if x < 1.:
        p2 = Wedge((x + 10., 1.5), 1., 45, -45, color='#FFFF00')
        ax.add_patch(p2)
    if x > 9:
        p2 = Wedge((x - 10., 1.5), 1., 45, -45, color='#FFFF00')
        ax.add_patch(p2)
    plt.show()
    
def closeMouth(x):
    fig, ax = plt.subplots(figsize=(10, 3), dpi=241.635, facecolor='#404040')
    ax.set_axis_off()
    ax.set_xlim(0., 10.)
    ax.set_ylim(-0.1, 3.1)
    plt.plot([0.25, 9.75], [2.75, 2.75], '#2121DE', linewidth=2)
    plt.plot([0.25, 9.75], [3.05, 3.05], '#2121DE', linewidth=2)
    plt.plot([0.25, 9.75], [0.25, 0.25], '#2121DE', linewidth=2)   
    plt.plot([0.25, 9.75], [-0.05, -0.05], '#2121DE', linewidth=2)   
    p = Circle((x, 1.5), 1., color='#FFFF00')
    ax.add_patch(p)
    if x < 1.:
        p2 = Circle((x + 10., 1.5), 1., color='#FFFF00')
        ax.add_patch(p2)
    if x > 9:
        p2 = Circle((x - 10., 1.5), 1., color='#FFFF00')
        ax.add_patch(p2)
    plt.show()
```
