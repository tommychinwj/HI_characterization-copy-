```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def connectivity1(i=0):
    
    one_frac = np.linspace(1., 2., 2)

    # prepare some coordinates
    x, y, z = np.indices((5, 5, 5))

    # draw cuboids in the top left and bottom right corners, and a link between them
    cube0 = (x >= 2) & (y >= 2) & (z >= 2) & (x < 3) & (y < 3) & (z < 3)
    cube1 = (x >= 2) & (y >= 2) & (z >= 2 + one_frac[i]) & (x < 3) & (y < 3) & (z < 3 + one_frac[i])
    cube2 = (x >= 2) & (y >= 2) & (z >= 2 - one_frac[i]) & (x < 3) & (y < 3) & (z < 3 - one_frac[i])
    cube3 = (x >= 2) & (y >= 2 + one_frac[i]) & (z >= 2) & (x < 3) & (y < 3 + one_frac[i]) & (z < 3)
    cube4 = (x >= 2) & (y >= 2 - one_frac[i]) & (z >= 2) & (x < 3) & (y < 3 - one_frac[i]) & (z < 3)
    cube5 = (x >= 2 + one_frac[i]) & (y >= 2) & (z >= 2) & (x < 3 + one_frac[i]) & (y < 3) & (z < 3)
    cube6 = (x >= 2 - one_frac[i]) & (y >= 2) & (z >= 2) & (x < 3 - one_frac[i]) & (y < 3) & (z < 3)
    cube7 = (x >= 2) & (y >= 2 + one_frac[i]) & (z >= 2 + one_frac[i]) & (x < 3) & (y < 3 + one_frac[i]) & (z < 3 + one_frac[i])
    cube8 = (x >= 2) & (y >= 2 - one_frac[i]) & (z >= 2 + one_frac[i]) & (x < 3) & (y < 3 - one_frac[i]) & (z < 3 + one_frac[i])
    cube9 = (x >= 2) & (y >= 2 + one_frac[i]) & (z >= 2 - one_frac[i]) & (x < 3) & (y < 3 + one_frac[i]) & (z < 3 - one_frac[i])
    cube10 = (x >= 2) & (y >= 2 - one_frac[i]) & (z >= 2 - one_frac[i]) & (x < 3) & (y < 3 - one_frac[i]) & (z < 3 - one_frac[i])
    cube11 = (x >= 2 + one_frac[i]) & (y >= 2 + one_frac[i]) & (z >= 2) & (x < 3 + one_frac[i]) & (y < 3 + one_frac[i]) & (z < 3)
    cube12 = (x >= 2 - one_frac[i]) & (y >= 2 + one_frac[i]) & (z >= 2) & (x < 3 - one_frac[i]) & (y < 3 + one_frac[i]) & (z < 3)
    cube13 = (x >= 2 + one_frac[i]) & (y >= 2 - one_frac[i]) & (z >= 2) & (x < 3 + one_frac[i]) & (y < 3 - one_frac[i]) & (z < 3)
    cube14 = (x >= 2 - one_frac[i]) & (y >= 2 - one_frac[i]) & (z >= 2) & (x < 3 - one_frac[i]) & (y < 3 - one_frac[i]) & (z < 3)
    cube15 = (x >= 2 + one_frac[i]) & (y >= 2) & (z >= 2 + one_frac[i]) & (x < 3 + one_frac[i]) & (y < 3) & (z < 3 + one_frac[i])
    cube16 = (x >= 2 + one_frac[i]) & (y >= 2) & (z >= 2 - one_frac[i]) & (x < 3 + one_frac[i]) & (y < 3) & (z < 3 - one_frac[i])
    cube17 = (x >= 2 - one_frac[i]) & (y >= 2) & (z >= 2 + one_frac[i]) & (x < 3 - one_frac[i]) & (y < 3) & (z < 3 + one_frac[i])
    cube18 = (x >= 2 - one_frac[i]) & (y >= 2) & (z >= 2 - one_frac[i]) & (x < 3 - one_frac[i]) & (y < 3) & (z < 3 - one_frac[i])
    cube19 = (x >= 2 + one_frac[i]) & (y >= 2 + one_frac[i]) & (z >= 2 + one_frac[i]) & (x < 3 + one_frac[i]) & (y < 3 + one_frac[i]) & (z < 3 + one_frac[i])
    cube20 = (x >= 2 - one_frac[i]) & (y >= 2 + one_frac[i]) & (z >= 2 + one_frac[i]) & (x < 3 - one_frac[i]) & (y < 3 + one_frac[i]) & (z < 3 + one_frac[i])
    cube21 = (x >= 2 + one_frac[i]) & (y >= 2 - one_frac[i]) & (z >= 2 + one_frac[i]) & (x < 3 + one_frac[i]) & (y < 3 - one_frac[i]) & (z < 3 + one_frac[i])
    cube22 = (x >= 2 + one_frac[i]) & (y >= 2 + one_frac[i]) & (z >= 2 - one_frac[i]) & (x < 3 + one_frac[i]) & (y < 3 + one_frac[i]) & (z < 3 - one_frac[i])
    cube23 = (x >= 2 - one_frac[i]) & (y >= 2 - one_frac[i]) & (z >= 2 + one_frac[i]) & (x < 3 - one_frac[i]) & (y < 3 - one_frac[i]) & (z < 3 + one_frac[i])
    cube24 = (x >= 2 + one_frac[i]) & (y >= 2 - one_frac[i]) & (z >= 2 - one_frac[i]) & (x < 3 + one_frac[i]) & (y < 3 - one_frac[i]) & (z < 3 - one_frac[i])
    cube25 = (x >= 2 - one_frac[i]) & (y >= 2 + one_frac[i]) & (z >= 2 - one_frac[i]) & (x < 3 - one_frac[i]) & (y < 3 + one_frac[i]) & (z < 3 - one_frac[i])
    cube26 = (x >= 2 - one_frac[i]) & (y >= 2 - one_frac[i]) & (z >= 2 - one_frac[i]) & (x < 3 - one_frac[i]) & (y < 3 - one_frac[i]) & (z < 3 - one_frac[i])

    
    # combine the objects into a single boolean array
    voxels = cube0 | cube1 | cube2 | cube3 | cube4 | cube5 | cube6 | cube7 | cube8 | cube9 | cube10 | cube11 | cube12 | cube13 | cube14 | cube15 | cube16 | cube17 | cube18 | cube19 | cube20 | cube21 | cube22 | cube23 | cube24 | cube25 | cube26

    # set the colors of each object
    colors = np.empty(voxels.shape, dtype=object)
    colors[cube0] = 'C0'
    colors[cube1] = 'C1'
    colors[cube2] = 'C1'
    colors[cube3] = 'C1'
    colors[cube4] = 'C1'
    colors[cube5] = 'C1'
    colors[cube6] = 'C1'
    colors[cube7] = 'C2'
    colors[cube8] = 'C2'
    colors[cube9] = 'C2'
    colors[cube10] = 'C2'
    colors[cube11] = 'C2'
    colors[cube12] = 'C2'
    colors[cube13] = 'C2'
    colors[cube14] = 'C2'
    colors[cube15] = 'C2'
    colors[cube16] = 'C2'
    colors[cube17] = 'C2'
    colors[cube18] = 'C2'
    colors[cube19] = 'C2'
    colors[cube20] = 'C2'
    colors[cube21] = 'C2'
    colors[cube22] = 'C2'
    colors[cube23] = 'C2'
    colors[cube24] = 'C2'
    colors[cube25] = 'C2'
    colors[cube26] = 'C2'





    # and plot everything
    fig = plt.figure(dpi=335.5, facecolor='#404040')
    ax = fig.gca(projection='3d')
    ax.set_facecolor('#404040')
    ax.voxels(voxels, facecolors=colors, edgecolor='k')
    ax.set_axis_off()
#     ax.view_init(0., 0.)

    plt.show()
```
